# Steps to get the sales for a certain day:
1. copy and paste the whole table part into a txt file (notepad or something)
2. change the path to the place where the above txt file is saved
3. change the dateSearch to the date you're looking for 

All the changes that must be made have capital letter comments next to them.

```python
import os
import re
path = "C:\\Users\\paige\\OneDrive\\Documents\\FasGas\\LiquorProductPriceChanges" #CHANGE THIS TO APPROPRIATE PATH
os.chdir(path)
files = os.listdir()
files
```

```python
dateSearch = "2019-05-10" #CHANGE THIS TO REFLECT THE FORMATTED DATE YOU WANT TO SEARCH

for file in files: #in case there's ever more than one file you want to parse, we'll start with this loop
    f = open(file, 'r') #open the file and call it "f"
    liquorStr = f.read() #read what's in the "f"ile
    liquorLines = liquorStr.split("\n") #split the read string into lines (\n = newline character)
    regex = re.compile("\s-[0-9]") #regular expression search for a sale (negative # in cost column)
    sales = list(filter(regex.search, liquorLines)) #search through liquorLines for the regex string, return a list
    certainDate = [d for d in sales if dateSearch in d] #return all instances where the date is the date you want
    for element in certainDate: #loop through and 
        print(element) #pretty print each element
```
